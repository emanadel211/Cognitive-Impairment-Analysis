```r
data = read.table("https://media.githubusercontent.com/media/ahmedmoustafa/gene-expression-datasets/main/datasets/cognitive_aging/cognitive_aging.tsv", header = TRUE, row.names = 1)
head(data)
```

```r
data = as.matrix(data)
head(data)
```

```r
dim(data)
```

```r
number_of_genes = nrow(data) # number of genes = number of rows
number_of_genes
```

```r
data2=log2(data+1)
hist(data2)
```

```r
colors = c(rep("blue", 9), rep("green",9) , rep("red",9 ), rep("yellow",9), rep("orange",13))
boxplot(data2, col = colors)
```

```r
#splitting data matrix into 5 groups
mon3 = data2[, 1:9]
head(mon3)
```

```r
mon6 = data2[, 10:18]
head(mon6)
```

```r
mon9= data2[, 19:27]
head(mon9)
```

```r
mon12= data2[, 28:36]
head(mon12)
```

```r
mon23= data2[, 37:49]
head(mon23)
```

```r
pvalue = NULL

for(i in 1:number_of_genes) {
  a = mon3[i, ]
  b = mon6[i, ]
  c = mon9[i, ]
  d = mon12[i, ]
  e = mon23[i, ]
  values = c(a, b, c, d, e)
  groups = c(rep("a", length(a)), rep("b", length(b)), rep("c", length(c)), rep("d", length(d)), rep("e", length(e)))
  anova_test = aov(values~groups)
  pvalue[i] = summary(anova_test)[[1]][["Pr(>F)"]][1]
}

head (pvalue)
```

```r
length(pvalue)
```

```r
hist(pvalue)
```

```r
hist(-log10(pvalue))
```

```r
#most significant differential expression or substantial fold change
```

```r
hist(-log(pvalue, base = 5))
```

```r
pvalue_cutoff = 0.05
```

```r
filter_by_pvalue_anova = pvalue <= pvalue_cutoff #in the paper it was 923
sum(filter_by_pvalue_anova)
```

```r
head(as.data.frame(filter_by_pvalue_anova))
```

```r
filtered_anova = data2[filter_by_pvalue_anova, ]
dim(filtered_anova)
```

```r
head (filtered_anova)
```
